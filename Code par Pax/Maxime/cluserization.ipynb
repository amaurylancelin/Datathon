```python
from Environnement.extractClusters import get_closest_keys_scoring, score_fn, get_cluster
```

```python
def fill_submission(df_submission, df_preds, rule="max"):
    """
    Fill the submission with the predictions of the model
    """
    df_submission["Cluster"] = df_submission.apply(lambda x: get_cluster(get_closest_keys_scoring(x["key"], df_preds, score_fn=score_fn), rule=rule), axis=1)
    return df_submission
```

```python
df_submission = fill_submission(df_submission, df_preds, rule="max")
```
